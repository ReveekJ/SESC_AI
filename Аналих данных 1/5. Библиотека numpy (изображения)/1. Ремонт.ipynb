```python
import numpy as np
from PIL import Image


def repair(filename: str):
    pic = np.array(Image.open(filename))
    back = pic[0, 0]
    new = np.where(pic == back, pic, 255 - pic)

    img = Image.fromarray(new)
    img.save('result.png')
    return img
```

```python
repair('../../data/ремонт.png')
```
